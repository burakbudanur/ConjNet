```python
import numpy as np  
from matplotlib import pyplot as plt 
import matplotlib.patches as mpatches


plt.rcParams.update({
    "text.usetex": True,
    "font.size": '12',
})
```

```python
fig_width = 19  # cm
fig_height = 5  # cm
fig = plt.figure(figsize=(fig_width / 2.54, fig_height / 2.54))
ax = fig.add_axes((0, 0, 1, 1))
ax.set_xlim(0, 1.0)
ax.set_ylim(-0.41, 0.31)
ax.set_axis_off()
arrow_props = {
    "width":1e-3, "head_length":2e-2, "head_width":3e-2, "length_includes_head":True, 
    "edgecolor":'black', "facecolor":'black',
    }

varrow_props = {
    "width":1e-3, "head_length":3e-2, "head_width":2e-2, "length_includes_head":True, 
    "edgecolor":'black', "facecolor":'black',
    }

rect_props = {
    "edgecolor":'black', "facecolor":'navajowhite',    
}
xscale = 1.0 / 1.1

ax.arrow(
    0 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(0.05 * xscale, 3e-2, "$u(t)$", horizontalalignment="center")
# ax.text(0.05 * xscale, -3e-2, "($30d$)", horizontalalignment="center")

ax.add_patch(
    mpatches.Rectangle((0.1 * xscale, -0.2), 0.1 * xscale, 0.4, **rect_props)
    )
ax.text(0.15 * xscale, 0.0, "$S$ \n (symm. \n red.)"
        ,horizontalalignment="center", 
        verticalalignment="center")

ax.arrow(
    0.2 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(0.25 * xscale, 3e-2, "$\\xi(t)$", horizontalalignment="center")
# ax.text(0.35 * xscale, -3e-2, "($29d$)", horizontalalignment="center")

# Trapez
x = np.array([0.3,0.3,0.4,0.4]) * xscale
y = [-0.3,0.3,0.1,-0.1]
ax.add_patch(
    mpatches.Polygon(
        xy=list(zip(x,y)), 
        fill=True, facecolor='navajowhite', edgecolor='black'
        )
        )
ax.text(0.35 * xscale, 0.0, "$E$ \n (enc.)"
        ,horizontalalignment="center", 
        verticalalignment="center")
ax.arrow(
    0.4 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(0.45 * xscale, 3e-2, "$\\eta(t)$", horizontalalignment="center")

ax.add_patch(
    mpatches.Rectangle((0.5 * xscale, -0.2), 0.1 * xscale, 0.4, **rect_props)
    )
ax.text(0.55 * xscale, 0.0, "$\mathcal{L}^{\\tau} $ \n (linear \n evol.)"
        ,horizontalalignment="center", 
        verticalalignment="center")

ax.arrow(
    0.6 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(0.65 * xscale, 3e-2, "$\\eta(t+\\tau)$", horizontalalignment="center")

# Trapez
x = np.array([0.7,0.7,0.8,0.8]) * xscale
y = [-0.1,0.1,0.3,-0.3]
ax.add_patch(
    mpatches.Polygon(
        xy=list(zip(x,y)), fill=True, facecolor='navajowhite', edgecolor='black'
        )
    )
ax.text(0.75 * xscale, 0.0, "$D$ \n (dec.)"
        ,horizontalalignment="center", 
        verticalalignment="center")
ax.arrow(
    0.8 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(0.85 * xscale, 3e-2, "$\\xi'(t+\\tau)$", horizontalalignment="center")

ax.add_patch(
    mpatches.Rectangle((0.9 * xscale, -0.2), 0.1 * xscale, 0.4, **rect_props)
    )
ax.text(0.95 * xscale, 0.0, "$A $ \n (symm. \n align.)"
        ,horizontalalignment="center", 
        verticalalignment="center")

ax.arrow(
    1.0 * xscale, 0, 0.1 * xscale, 0.0, **arrow_props
)
ax.text(1.05 * xscale, 3e-2, "$u'(t+\\tau)$", horizontalalignment="center")

ax.plot(np.array([0.05, 0.05]) * xscale, [0.0, -0.35], color='black')
ax.plot(np.array([0.05, 0.95]) * xscale, [-0.35, -0.35], color='black')
ax.arrow(0.95 * xscale, -0.35, 0.0, 0.15, **varrow_props)

# fig.savefig("architecture.eps")
fig.savefig("../tex/fig/architecture.png", dpi=200)

# ax.text(0.55 * xscale, -3e-2, "($3d$)", horizontalalignment="center")

# ax.tick_params(bottom=False, top=False, left=False, right=False)
# ax.tick_params(labelbottom=False, labeltop=False, labelleft=False, labelright=False)
```
